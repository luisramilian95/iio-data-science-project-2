## **Carga de datos**

Se cargan los datos para este trabajo

## **Pregunta #5**

¿Cuál es la probabilidad de obtener, por booster, una carta rara, súper rara, ultra rara o secreta?

Para iniciar este problema, es necesario saber cómo se distribuye un booster:\
\
Todo booster posee 9 cartas. 8 de ellas son cartas comunes al azar, y 1 es una carta que puede ser rara, súper rara, ultra rara o secreta.\
\
Para poder llevar a cabo esto, se supuso que la tasa de impresión de las cartas es la misma. Esto hace que la probabilidad de una carta no común siempre es 1/9.\
\
Establecido esto, se determinó la proporción de cartas de distinta rareza en cada booster mediante un histograma.

```python
df_card_rarity=dataset[['booster_name','card_rarity','card_number']].groupby(['booster_name','card_rarity']).count().reset_index()
fig = px.histogram(df_card_rarity, x="card_rarity", y="card_number",
             color='booster_name', barmode='group',
             height=400)
fig.show()
```

```python
df_card_rarity_filter=dataset.query("card_rarity=='Secret Rare' or card_rarity=='Super Rare' or card_rarity=='Rare' or card_rarity=='Ultra Rare' or card_rarity=='Ultimate Rare'").groupby(['booster_name','card_rarity']).count().reset_index()
fig = px.histogram(df_card_rarity_filter, x="card_rarity", y="card_number",
             color='booster_name', barmode='group',
             height=400)
fig.show()
```

Filtrando las cartas por rareza, se obtiene la proporción de ellas por booster. Esta proporción se multiplica por 1/9 para obtener las probabilidades que se muestran en la próxima tabla

```python
df_card_rarity_filter[['booster_name','card_rarity','card_number']].sort_values(['booster_name','card_number','card_rarity'])
```

```python
booster_card_rarity = df_card_rarity_filter.groupby(['booster_name','card_rarity']).agg({'card_number': 'sum'})
booster = df_card_rarity_filter.groupby(['booster_name']).agg({'card_number': 'sum'})
booster_rare_prop=booster_card_rarity.div(booster, level='booster_name') *1/9
```

```python
booster_rare_prop.sort_values(['booster_name','card_number'])
```
